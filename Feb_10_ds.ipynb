```python
# def function


def sqr(num):
   result = num**2
   return result
```

```python
sqr(10)
```

```python
def num_status(n):
  if n>0:
    if n%2==0:
      return "Even"
    else:
      return "Odd"
  else:
      return "Low value, Enter the number again!"
```

```python
print(num_status(5))
```

```python
# ques: create a function to return a proper table of given number.

def multiplication_table(num):
    print(f"Multiplication Table of {num}")
    print("-" * 30)
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")
```

```python
multiplication_table(5)
```
