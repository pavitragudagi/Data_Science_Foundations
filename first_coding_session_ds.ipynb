```python
# practice question
# take input from user.
# if the num is even then get the square of it.
# if the num is  odd then multiply it with 10.
# if the num is zero then add 500 in it.



user_input = int(input("Enter a number: "))

if  user_input%2==0 and user_input>0:
  print("Number is even")
  print(f"The square of {user_input} is {user_input**2}")
elif user_input%2!=0:
  print("Number is odd")
  print(f"The multiplication of {user_input} is {user_input*10}")
elif user_input==0:
  print("Number is zero")
  print(f"The addition of {user_input} is {user_input+500}")


```
