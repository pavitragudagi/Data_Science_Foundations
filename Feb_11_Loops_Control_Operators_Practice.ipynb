```
# Arithmetic Operators Example
a = 10
b = 3

print('Addition:', a + b)
print('Subtraction:', a - b)
print('Multiplication:', a * b)
print('Division:', a / b)
print('Floor Division:', a // b)
print('Exponent:', a ** b)
```

```
# Relational and Logical Operators
x = 5
y = 8

print('x > y:', x > y)
print('x < y and y > 0:', x < y and y > 0)
print('x == 5 or y == 0:', x == 5 or y == 0)
```

```
# For Loop Example
for i in range(1, 6):
    print('Iteration:', i)
```

```
# While Loop Example
count = 1
while count <= 5:
    print('Count:', count)
    count += 1
```

```
# Break and Continue Example
for num in range(1, 10):
    if num == 5:
        break
    print('Number:', num)

for num in range(1, 6):
    if num == 3:
        continue
    print('After Continue:', num)
```
