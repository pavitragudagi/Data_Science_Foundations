```python
for i in range(50, 100):
  if i%17==0 or i%23==0:
    print("Buzz")
  else:
    print("Num")
```

```python
for i in range(50, 100):
  if i%4==0 or i%8==0:
    print("Buzz")
  elif i%5==0 or i%10==0:
    print("Fizz")
  elif i%11==0:
    print("Tom")
  else:
    print(i)
```

```python
num = int(input("Enter a number: "))


while True:
  if num > 10:
    print(num+10)
    break
  else:
     print("Please enter the number greater than 10.")
```

```python
while True:
  try:
    num = int(input("Enter a number: "))
    if num > 10:
      print(num+10)
      break
    else:
      print("Low digit, try again!")
  except ValueError:
    print("Please try with a valid number. ")
```
