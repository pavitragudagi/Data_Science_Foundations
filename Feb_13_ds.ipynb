# 📘 Advanced String and List Operations in Python
This notebook covers detailed string and list operations with multiple examples.


## 🔹 PART 1: STRING OPERATIONS


```python
# Creating Strings
str1 = "Hello"
str2 = 'Python'
str3 = """Multi-line
String Example"""
print(str1)
print(str2)
print(str3)
```

```python
# Concatenation and Repetition
a = "Data"
b = "Science"
print(a + " " + b)
print(a * 3)
```

```python
# Indexing and Slicing
word = "Programming"
print(word[0])
print(word[-1])
print(word[0:6])
print(word[::-1])  # Reverse string
```

```python
# String Methods
text = " python programming language "
print(text.upper())
print(text.lower())
print(text.title())
print(text.strip())
print(text.replace("python", "Java"))
print(text.split())
```

```python
# Checking String Properties
sample = "Python123"
print(sample.isalpha())
print(sample.isdigit())
print(sample.isalnum())
print(sample.startswith("Py"))
print(sample.endswith("123"))
```

```python
# String Formatting
name = "Pavitra"
age = 21
print("My name is {} and I am {} years old".format(name, age))
print(f"My name is {name} and I am {age} years old")
```

## 🔹 PART 2: LIST OPERATIONS


```python
# Creating Lists
numbers = [10, 20, 30, 40]
mixed = [1, "Python", 3.5, True]
print(numbers)
print(mixed)
```

```python
# Indexing and Slicing
print(numbers[0])
print(numbers[-1])
print(numbers[1:3])
```

```python
# Adding Elements
numbers.append(50)
numbers.insert(1, 15)
print(numbers)
```

```python
# Removing Elements
numbers.remove(30)
numbers.pop()
print(numbers)
```

```python
# Sorting and Reversing
nums = [5, 2, 9, 1]
nums.sort()
print("Sorted:", nums)
nums.reverse()
print("Reversed:", nums)
```

```python
# List Functions
print("Length:", len(nums))
print("Max:", max(nums))
print("Min:", min(nums))
print("Sum:", sum(nums))
```

```python
# Looping Through List
for item in nums:
    print(item)
```

```python
# List Comprehension
squares = [x**2 for x in range(1, 11)]
even_numbers = [x for x in range(1, 21) if x % 2 == 0]
print("Squares:", squares)
print("Even Numbers:", even_numbers)
```

## 🎯 Practice Exercises
1. Reverse a string without using slicing.
2. Find the second largest number in a list.
3. Count vowels in a given string.


# ✅ Solutions to Practice Exercises


```python
# 1️⃣ Reverse a string without using slicing
text = "Python"
reversed_text = ""
for char in text:
    reversed_text = char + reversed_text
print("Reversed String:", reversed_text)
```

```python
# 2️⃣ Find the second largest number in a list
numbers = [10, 45, 32, 67, 89, 23]
unique_numbers = list(set(numbers))
unique_numbers.sort()
second_largest = unique_numbers[-2]
print("Second Largest Number:", second_largest)
```

```python
# 3️⃣ Count vowels in a given string
sentence = "Data Science"
vowels = "aeiouAEIOU"
count = 0
for char in sentence:
    if char in vowels:
        count += 1
print("Number of vowels:", count)
```
